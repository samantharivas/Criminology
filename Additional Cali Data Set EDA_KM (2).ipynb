```python
import matplotlib.pyplot as plt
```

```python
# counts for race
race_counts = ca_df['Subjects_race'].value_counts()
officer_involved_counts = ca_df[ca_df['Intentional_Use_of_Force_(Developing)'] == 1]['Subjects_race'].value_counts()

plt.figure(figsize=(20, 15))
plt.bar(race_counts.index, race_counts.values, label='Total Fatalities')
plt.bar(officer_involved_counts.index, officer_involved_counts.values, label='Intentional_Use_of_Force_(Developing)', alpha=0.7)
plt.xlabel('Subjects_race')
plt.ylabel('Count')
plt.title('Fatalities by Race')
plt.legend()
plt.show()
```

```python
# Cause of Death Counts 
cause_counts = ca_df['Cause_of_death'].value_counts()
officer_involved_counts = ca_df[ca_df['Intentional_Use_of_Force_(Developing)'] == 1]['Cause_of_death'].value_counts()

plt.figure(figsize=(30, 20))
plt.bar(race_counts.index, race_counts.values, label='Total Fatalities')
plt.bar(officer_involved_counts.index, officer_involved_counts.values, label='Intentional_Use_of_Force_(Developing)', alpha=0.7)
plt.xlabel('Cause_of_death')
plt.ylabel('Count')
plt.title('Fatalities by Cause of Death')
plt.legend()
plt.show()
```

```python
# Agency Counts
agency_counts = ca_df['Agency_responsible_for_death'].value_counts().sort_values(ascending=False)
agency_counts
```

```python
# Agency Counts
agency_counts = ca_df['Agency_responsible_for_death'].value_counts().sort_values(ascending=False)

# Filter agencies with more than 5 cases
filtered_agencies = agency_counts[agency_counts > 15]

# Filter the original DataFrame to include only the selected agencies
filtered_df = ca_df[ca_df['Agency_responsible_for_death'].isin(filtered_agencies.index)]

# Plot the bar graph
plt.figure(figsize=(12, 8))
plt.bar(filtered_agencies.index, filtered_agencies.values)
plt.xlabel('Agency_responsible_for_death')
plt.ylabel('Count')
plt.title('Fatalities by Agency (Cases > 15)')
plt.xticks(rotation=90) 

# Add labels to the bars
for i, count in enumerate(filtered_agencies.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

```python
# Agency Counts
agency_counts = ca_df['Agency_responsible_for_death'].value_counts().sort_values(ascending=False)

# Filter agencies with more than 5 cases
filtered_agencies = agency_counts[(agency_counts > 5) & (agency_counts <= 15)]

# Filter the original DataFrame to include only the selected agencies
filtered_df = ca_df[ca_df['Agency_responsible_for_death'].isin(filtered_agencies.index)]

# Plot the bar graph
plt.figure(figsize=(12, 8))
plt.bar(filtered_agencies.index, filtered_agencies.values)
plt.xlabel('Agency_responsible_for_death')
plt.ylabel('Count')
plt.title('Fatalities by Agency (Cases > 5 and <= 15)')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability

# Add labels to the bars
for i, count in enumerate(filtered_agencies.values):
    plt.text(i, count, str(count), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
